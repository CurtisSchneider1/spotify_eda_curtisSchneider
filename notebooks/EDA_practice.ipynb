Exploratory Data Analysis - Process of Understanding the Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/dataset.csv")

print(df.shape)

df.head(10)
```

```python
df.columns
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
# isnull is just an alias of isna
# True = NA
df.isna()

df.isna().sum()
```

```python
# Just to show that isnull returns exact same results
df.isnull().sum()
```

```python
df.loc[df['artists'].isna()]
```

```python
# keep parameter - values are first last and false, set to first by default
# False returns True on any duped row
# first is the first duped row
# last is the last duped row in the dataset, regardless of how many dupes before it
df.duplicated(keep=False).sum()
```

```python
# sort_values to put the dupes next to each other
df.loc[df.duplicated(keep=False)].sort_values("track_id")
```

```python
# checks for just dupes in track_id column
df.loc[df["track_id"].duplicated(keep=False)].sort_values("track_id")
```

```python
# univariate analysis = looking at individual columns
# qualitative (categorical) variables - like text data that can't be counted or measured
# - we can get a count of unique values
# informs us how about how these variables need to be encoded before going to a model
df["artists"].value_counts()
```

```python
# quantitative variables - can be counted or measured
df.describe()

# 50% quartile is the median
```

```python
df["danceability"].hist(bins=60)
```

```python
sns.boxplot(x=df["danceability"])
```

```python
# multivariate analysis - variables looks at how 2 or more variables are related to each other
# corr only works on numerical data
df.columns
df_corr = df[['popularity',
       'duration_ms', 'danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'time_signature']].corr()

df_corr
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(df_corr, annot=True, cmap='mako')       # cmap is color palette

# look for negatively and positively correlated values
```

```python
# pair plot show relationship between 2 variables on a scatter plot
# -also contains a histogram which show the distribution of each variable
sns.pairplot(df[["acousticness", "energy"]], height=5, aspect=1.5)
```

```python
# let's take a random sample of the df with fewer points
df_sample = df[["acousticness", "energy"]].sample(n=5000, random_state=1)   # n is how many data points, random_state gives reproducable results
sns.pairplot(df_sample, height=5, aspect=1.5, diag_kind="kde")              # diag_kind changes the bar chart to specified chart
```

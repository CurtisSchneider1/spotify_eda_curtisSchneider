Exploratory Data Analysis - Process of Understanding the Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/dataset.csv")

print(df.shape)

df.head(10)
```

```python
df.columns
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
# isnull is just an alias of isna
# True = NA
df.isna()

df.isna().sum()
```

```python
# Just to show that isnull returns exact same results
df.isnull().sum()
```

```python
df.loc[df['artists'].isna()]
```

```python
# keep parameter - values are first last and false, set to first by default
# False returns True on any duped row
# first is the first duped row
# last is the last duped row in the dataset, regardless of how many dupes before it
df.duplicated(keep=False).sum()
```

```python
# sort_values to put the dupes next to each other
df.loc[df.duplicated(keep=False)].sort_values("track_id")
```

```python
# checks for just dupes in track_id column
df.loc[df["track_id"].duplicated(keep=False)].sort_values("track_id")
```

```python
# univariate analysis = looking at individual columns
# qualitative (categorical) variables - like text data that can't be counted or measured
# - we can get a count of unique values
# informs us how about how these variables need to be encoded before going to a model
df["artists"].value_counts()
```

```python
# quantitative variables - can be counted or measured
df.describe()

# 50% quartile is the median
```
